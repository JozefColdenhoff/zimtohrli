```python
import sys
import os
import scipy
import numpy as np
import matplotlib.pyplot as plt
import IPython.display as ipd
sys.path.append(os.path.join("..", "..", "build"))
import pyohrli
```

```python
sig = scipy.signal.chirp(
    np.linspace(0.0, 0.2, 48000), 100, 0.2, 15000, method="logarithmic"
)
ipd.Audio(sig, rate=48000)
```

```python
p = pyohrli.Pyohrli()
out = p.analyze(sig)
plt.imshow(np.flip(out, axis=1).T)
plt.show()
```
